```python
import csv
import pandas as pd
from pathlib import Path

f = open('RP15_otter_ai.txt')
csv_f = csv.reader(f)
data = []
row = []
timestamp_From = ""
timestamp_To = ""
speaker = ""
transcription = ""

for row in csv_f:

    transcription = ""
    for item in row:
        if('speaker' in item.lower()):
            item = item.replace(" ", "")
            # Get the Timestamp From value assumed as 00:00 format
            timestamp_From = item[-4:]
            print(timestamp_From)
            speaker = item[:-4]
        else:
            transcription += item
    if transcription != '' and transcription != 'Transcribed by https://otter.ai':
        data.append(['0', timestamp_From, timestamp_To,
                    speaker, transcription])


# print(data)
df = pd.DataFrame(data, columns=["", "from", "to", "speaker", "transcript"])
print(df)

# Write out as new CSV file
filepath = Path('out.csv')

filepath.parent.mkdir(parents=True, exist_ok=True)
df.to_csv(filepath, index=False)
```
